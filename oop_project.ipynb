```python
class Exercise:
    def __init__(self, name, calories_burned):
        self.name = name
        self.calories_burned = calories_burned

class Workout:
    def __init__(self, name):
        self.name = name
        self.exercises = []

    def add_exercise(self, exercise):
        self.exercises.append(exercise)

    def calculate_calories_burned(self):
        total_calories = 0
        for exercise in self.exercises:
            total_calories += exercise.calories_burned
        return total_calories

class FitnessApp:
    def __init__(self):
        self.workouts = []

    def add_workout(self, workout):
        self.workouts.append(workout)

    def display_workouts(self):
        for i, workout in enumerate(self.workouts, start=1):
            print(f"Workout {i}: {workout.name}")

    def track_calories_burned(self):
        self.display_workouts()
        workout_index = int(input("Enter the workout number: ")) - 1

        if workout_index < 0 or workout_index >= len(self.workouts):
            print("Invalid workout number!")
            return

        workout = self.workouts[workout_index]
        print(f"\nTracking {workout.name}...\n")

        for i, exercise in enumerate(workout.exercises, start=1):
            print(f"Exercise {i}: {exercise.name}")
            calories_burned = exercise.calories_burned
            print(f"Calories burned: {calories_burned}\n")

        total_calories = workout.calculate_calories_burned()
        print(f"Total calories burned: {total_calories}")

# Create exercises
push_ups = Exercise("Push-ups", 100)
sit_ups = Exercise("Sit-ups", 75)
jumping_jacks = Exercise("Jumping Jacks", 50)

# Create workouts
workout1 = Workout("Morning Workout")
workout1.add_exercise(push_ups)
workout1.add_exercise(sit_ups)

workout2 = Workout("Evening Workout")
workout2.add_exercise(push_ups)
workout2.add_exercise(jumping_jacks)

# Create fitness app
fitness_app = FitnessApp()
fitness_app.add_workout(workout1)
fitness_app.add_workout(workout2)

# Track calories burned
fitness_app.track_calories_burned()
```
